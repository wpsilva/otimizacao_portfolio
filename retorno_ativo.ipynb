```python
import yfinance as yf
import matplotlib as plt
import pandas as pd
import numpy as np
```

```python
data_inicial = '2015-05-22'
ativos = ['ITSA4', 'ABEV3', 'WEGE3', 'EGIE3', 'VIIA3', 'OIBR3', 'PETR4','GGBR3', 'VILG11', 'CMIN3', 'MGLU3', 'VIVT3']
nome_ativos = (pd.Series(ativos)+'.SA').tolist()
```

```python
fechamentos = yf.download(tickers = nome_ativos, start = '2015-02-22', rounding = True)['Adj Close']
```

```python
retornos = fechamentos.pct_change()
```

```python
retornos.columns = retornos.columns.str.rstrip(".SA")
retornos.plot()
```

```python
rm= retornos.mean()
```
